### Notes

* General coordinate transformation used, i.e. to assign node geometry and advect the particles when periodic

$$ x = r \cos(\theta) \;\;\;\;\;\; y = r \sin(\theta)
   $$

* All equations in Underworld are still written and discretised using the Cartesian coordinate system. Only the nodes are laid out in a spherical/cylindrical geometry

* For free slip boundaries we require to rotate the boundary velocity degrees of freedom (dofs) to be normal and tangential to the surface. This is what the Spherical/Stokes.xml provides

* In the case of a full annulus with both walls having free-slip boundary conditions we have a stokes flow **null-space**. That is, a global velocity solution (solid body rotation flow) that the mechanical problem allows, it makes no difference to the governing equations/physics if the flow exists or not!!!
This is bad in two way:
    1. because our model can rotate unexpectantly. 
    2. the direct solver won't work because we can't invert the stiffness matrix K
    
    To avoid this null space problem - we find the null space vector in a post processing step after stokes system is solved and **remove it** from the velocity field.   
   
   This means a solid body rotation velocity solution is removed in models in Spherical Cylindrical coordinates.
   But, for models that we want to have a solid body rotation flow, i.e. lidriven model above, I left an option to disable the null-space removal, the option is 
   ~~~~
   --AllowSolidBodyRotation=True
   ~~~~

* Velocity output is still in Cartesian coordinates

* Multigrid and checkpointing in parallel have been working for me within Cylindrical geometry

# Redefinition of wall boundary conditions

The walls and specification of the velocity component vx,vy,vz still exist, i didn't rename them as the underlying code was to stupid to rename (in the new version of underworld we have resolved this issue of naming variables).
Instead I used the existing names of walls and components but I redefined their meanings. Below is a guid for the definitions.

## 2D cylindrical mesh

Walls

 * 'right'/'left' wall are perpendicular to OUTER/INNER radius vector 
 * 'bottom'  == clockwise facing out wall
 * 'top'     == anticlockwise facing out wall

  for a periodic annulus we set periodic_y == True, causing the top/bottom walls to become period
Components:

 * 'vx' is radial component
 * 'vy' is angular (anticlockwise) component

## 3D r,theta,phi mesh

Walls

 * 'right'/'left' wall are perpendicular to OUTER/INNER radius
 * 'top'/'bottom' wall are perpendicular to E-W 
 * 'front'/'back' wall are perpendicular to N-S 

Components

 * 'vx' is radial component
 * 'vy' is angular component along E-W walls. Note the vector is aligned with E-W direction
 * 'vz' is angular component along N-S walls. Note the vector is aligned with N-S direction
     
## 3D Regional mesh

Walls

 * 'right'/'left' walls are perpendicular to OUTER/INNER radius vector
 * 'top'/'bottom' walls are perpendicular to the great circles that go through E-W
 * 'front'/'back' walls are perpendicular to the great circles that go through N-S

Component

  * 'vx' is radial component
  * 'vy' is angular component along E-W walls. Note the vector is **NOT** aligned with E-W direction
  * 'vz' is angular component along N-S walls. Note the vector is aligned with N-S direction

# XML notes, the new stuff

##### Add Spherical toolbox in the top

~~~~
  <list name="import">
      <param>Spherical</param>
  </list>
~~~~
____ 
##### Swap out the old definition of the mesh

Previously called

~~~~
    <include> StgFEM/LinearMesh.xml </include>
~~~~

Is now either 

~~~~
    <include> Spherical/RSMesh.xml </include>
~~~~ 

for the Regional section or

~~~~
    <include> Spherical/SphericalMesh.xml </include> 
~~~~

for the cylindrical, r,theta,phi mesh

____
##### Redefinition of discretisation 

For the Spherical (cylinder or r,theta,phi) mesh

~~~
    minX / maxX --> the radial min / max
    minY / maxY --> the longitude min / max (in degrees)
    minY / maxY --> the latitude min / max (in degrees)
~~~

For the Regional section mesh

~~~
    minX / maxX --> the radial min / max
    minY / maxY --> the angle of the two great circle in the  (in degrees)
    minY / maxY --> the theta min / max (in degrees)
~~~
 
____
##### The new definition of gravity, it is now a vector that is evaluate for every point individually

~~~
    <struct name="bodyForce">
      <param name="Type">Ppc_PointGravity</param>
      <param name="Alpha">DensityLabel</param>
      <param name="xcoord">0.0</param>
      <param name="ycoord">0.0</param>
      <param name="zcoord">0.0</param>
    </struct>
~~~

 * The xcoord, ycoord, zcoord defines the point location in model space that gravity pulls things toward.
 * The Alpha is a function that determines the magnitude of the gravity. If you want a depth dependant gravity Alpha must control the dependence as the xcoord, ycoord, zcoord and only used to determine a unit vector direction for the gravity

_____
##### New particle distribution based on per element approach

This method addds the 'cellParticleCount' number of particles to each cell, randomly generating the coordinate in the cell's local coordinates. This is unlike the previous method which used global domain coordinate to generate particle coordinates.
For a spherical/cyclindrical mesh this means you'll have a higher density of particles towards the inner of the domain.

~~~
    <struct name="materialSwarmParticleLayout" mergeType="merge">
      <param name="Type" mergeType="replace">IrregularMeshParticleLayout</param>
      <param name="cellParticleCount">20</param>
      <param name="FeMesh">linearMesh</param>
    </struct>
~~~


 



# The Cylindrical Benchmarks

Rhodri Davies lead a cylindrical convection benchmark paper, as yet unpublished. The benchmark problems are all in 2D annulus domains and cover a range of:

 * convection equation approximations; ALA, TALA, EBA and BA
 * boundary conditions
 * flow control parametrisations

Underworld's contribution is located under the directory

~~~
$UW_DIR/libUnderworld/Spherical/CylindricalBenchmarks/inputfiles/
~~~

All models are run using python scripts. Note for bm_4, a mismatch in the Nusselt number appears to be a problem that is unresolved. More work on this is needed
